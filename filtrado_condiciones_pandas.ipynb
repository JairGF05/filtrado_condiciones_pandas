```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep = ',', header = 0)
df_books.head(2)
```

```python
#Mostrar en logica Booleana quienes cumplen con tener un año mayor a 2016
df_books['Year'] > 2016
```

```python
mayor_a_2016 = df_books['Year'] > 2016
```

```python
#Pasarlo a un dataframe
df_books[mayor_a_2016]
```

```python
#Obtener registros con el genero de tipo Fiction
df_books['Genre'] == 'Fiction'
```

```python
libros_ficcion = df_books['Genre'] == 'Fiction'
```

```python
df_books[libros_ficcion]
```

```python
#libros de ficcion publicados despues del 2016
df_books[mayor_a_2016 & libros_ficcion]
```

```python
#btener datos con años menor o igual a 2016 usando negación
df_books[-mayor_a_2016]
```
